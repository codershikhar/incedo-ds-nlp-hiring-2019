```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train_file.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(columns=['GeoLocation'], inplace=True)
```

```python
df.describe(include='all')
```

```python
Greater_Risk_Question = df.Greater_Risk_Question.unique()
Greater_Risk_Question
```

```python
for i in Greater_Risk_Question:
    print(len(df.loc[df.Greater_Risk_Question == i, 'Description'].unique()))
    print(i)
    print(df.loc[df.Greater_Risk_Question == i, 'Description'].unique())
```

```python
df.drop(columns=['Description'], inplace=True)
```

```python
df.info()
```

```python
### label encoding
### LocationDesc, Greater_Risk_Question, Sex, Race, QuestionCode, StratificationType
df_encoded = df.copy()

from sklearn.preprocessing import LabelEncoder
encoders = {}
for clm in ['LocationDesc', 'Greater_Risk_Question', 'Sex', 'Race', 'QuestionCode', 'StratificationType']:
    le = LabelEncoder()
    df_encoded[clm] = le.fit_transform(df_encoded[clm])
    encoders[clm] = le
df_encoded.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
df_scaled = df_encoded.copy()
scaler = {}
for clm in ['YEAR', 'LocationDesc', 'Subtopic', 'Greater_Risk_Question', 'Sample_Size', 'Sex', 'Race', 
            'Grade', 'QuestionCode', 'StratID1', 'StratID2', 'StratID3', 'StratificationType']:
    sc = MinMaxScaler()
    df_scaled[clm] = sc.fit_transform(df_scaled[clm].values.reshape(-1, 1))
    scaler[clm] = sc
```

```python
df_scaled.head()
```

```python
X = df_scaled[['YEAR', 'LocationDesc', 'Subtopic', 'Greater_Risk_Question', 'Sample_Size', 'Sex', 'Race', 
            'Grade', 'QuestionCode', 'StratID1', 'StratID2', 'StratID3', 'StratificationType']]
Y = df_scaled[['Greater_Risk_Probability']]
```

```python
X.head()
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state=42)
```

```python
from sklearn.neural_network import MLPRegressor
nn1 = MLPRegressor(hidden_layer_sizes=(15,30,30,15,), verbose=True, max_iter=1500)
nn1
```

```python
# nn1.fit(X=X_train, y=y_train)
nn1.fit(X=X, y=Y)
```

```python
print('train:', nn1.score(X, Y))
print('train:', nn1.score(X_train, y_train))
print('test:', nn1.score(X_test, y_test))
```

```python
test = pd.read_csv('test_file.csv')
```

```python
test.head()
```

```python
test.drop(columns=['GeoLocation'], inplace=True)
test.drop(columns=['Description'], inplace=True)
```

```python
for clm in ['LocationDesc', 'Greater_Risk_Question', 'Sex', 'Race', 'QuestionCode', 'StratificationType']:
    test[clm] = encoders[clm].transform(test[clm])
test.head()
```

```python
for clm in ['YEAR', 'LocationDesc', 'Subtopic', 'Greater_Risk_Question', 'Sample_Size', 'Sex', 'Race', 
            'Grade', 'QuestionCode', 'StratID1', 'StratID2', 'StratID3', 'StratificationType']:
    test[clm] = scaler[clm].transform(test[clm].values.reshape(-1, 1))
test.head()
```

```python
prediction = pd.DataFrame()
prediction['Patient_ID'] = test['Patient_ID']
prediction['Greater_Risk_Probability'] = nn1.predict(test[['YEAR', 'LocationDesc', 'Subtopic', 'Greater_Risk_Question', 
                                                           'Sample_Size', 'Sex', 'Race', 'Grade', 'QuestionCode', 
                                                           'StratID1', 'StratID2', 'StratID3', 'StratificationType']])
```

```python
prediction.to_csv('sample_submission.csv', index=False)
prediction.head()
```
