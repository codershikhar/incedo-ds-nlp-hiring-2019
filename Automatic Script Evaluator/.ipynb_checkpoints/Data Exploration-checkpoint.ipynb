```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train_dataset.csv')
df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
### reaplace Nan Essayset with the previous value
for index, row in df.iterrows():
    if row.isnull()['Essayset']:
        df.loc[index, 'Essayset'] = df.loc[index-1].Essayset
df.info()
```

```python
### create a field avg_score
### ignore nan to canculate avg
df['avg_score'] = 0
for index, row in df.iterrows():
    total = 0
    div = 0
    if row.isna().score_1 == False:
        total += row.score_1
        div += 1
    if row.isna().score_2 == False:
        total += row.score_2
        div += 1
    if row.isna().score_3 == False:
        total += row.score_3
        div += 1
    if row.isna().score_4 == False:
        total += row.score_4
        div += 1
    if row.isna().score_5 == False:
        total += row.score_5
        div += 1
    df.loc[index, 'avg_score'] = np.round(total/div)
df.head()
```

```python
df[['score_1', 'score_2', 'score_3', 'score_4', 'score_5', 'avg_score']].corr()
```

```python
df.info()
```

```python
df_encoded = df.copy()

from sklearn.preprocessing import LabelEncoder
encoders = {}
for clm in ['clarity', 'coherent']:
    le = LabelEncoder()
    df_encoded[clm] = le.fit_transform(df_encoded[clm].astype(str))
    encoders[clm] = le
df_encoded.head()
```

```python
for i in df.Essayset.unique():
    print(i, max(df.loc[df.Essayset == i].groupby(['Essayset', 'clarity']).size()))
```

```python
### replace all nan in coherent and clarity
for index, row in df.iterrows():
    if row.isna().coherent == True:
        df.loc[index, 'coherent'] = list(list(dict(df.loc[df.Essayset == row.Essayset].groupby(['Essayset', 'coherent']).size().nlargest(1)).keys())[0])[1]
    if row.isna().clarity == True:
        df.loc[index, 'clarity'] = list(list(dict(df.loc[df.Essayset == row.Essayset].groupby(['Essayset', 'clarity']).size().nlargest(1)).keys())[0])[1]
df.info()
```

```python
X = df[['EssayText']]
Y = df[['avg_score']]
print(X.shape)
print(Y.shape)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tf = TfidfVectorizer(stop_words='english', min_df=2, ngram_range=(1, 3))
tf
```

```python
tf.fit(X.EssayText)
```

```python
tf.vocabulary_
```

```python
x_tf = tf.transform(X.EssayText)
x_tf.shape
```

```python
from sklearn.neural_network import MLPClassifier
nn = MLPClassifier(hidden_layer_sizes=(1000,100,10,), verbose=True)
nn
```

```python
nn.fit(x_tf, Y.avg_score)
```

```python
nn.score(x_tf, Y)
```

```python
test = pd.read_csv('test_dataset.csv')
```

```python
test.head()
```

```python
x_test = test.EssayText
```

```python
x_test_tf = tf.transform(x_test)
x_test_tf.shape
```

```python
test['essay_score'] = np.round(nn.predict(x_test_tf))
```

```python
test.head()
```

```python
pre = pd.DataFrame()
pre['id'] = test['ID']
pre['essay_set'] = test['Essayset']
pre['essay_score'] = test['essay_score']
pre.to_csv('submission.csv', index=False)
```
